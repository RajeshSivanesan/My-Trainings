```python
import pandas as pd
df = pd.read_csv('Predict Cab price.csv')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
df
```

```python
result = model.fit(df[['ncabs']], df['price'])
```

```python
result
```

```python
print('intercept:', result.intercept_)
```

```python
print('slope:', result.coef_)
```

```python
y_pred = result.intercept_ + model.coef_[0] * df['price']
```

```python
y_pred
```

```python
y_pred = model.predict(df[['ncabs']])
```

```python
y_pred
```

```python
actual_price = df['price']
```

```python
actual_price
```

```python
from sklearn.metrics import r2_score, mean_squared_error, mean_absolute_error
```

```python
r2_score(actual_price, y_pred)
```

```python
mean_squared_error(actual_price, y_pred)
```

```python
mean_absolute_error(actual_price, y_pred)
```

```python
import numpy as np
np.sqrt(mean_squared_error(actual_price, y_pred))
```

```python
import matplotlib.pyplot as plt
plt.scatter(df['ncabs'], df['price'])
plt.plot(df['ncabs'], y_pred)
plt.show()
```

```python
def mean_absolute_percentage_error(y_true, y_pred):
    y_true, y_pred = np.array(y_true), np.array(y_pred)
    return np.mean(np.abs((y_true - y_pred) / y_true)) * 100
```

```python
mean_absolute_percentage_error(actual_price, y_pred)
```
