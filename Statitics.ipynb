```python
import pandas as pd
```

```python
import matplotlib as plot
import seaborn as sea
import matplotlib.pyplot as plt
```

```python
sea.set(rc={'figure.figsize': (12, 12)})
sea.barplot(data = dataFrame1)
```

```python
import matplotlib.pyplot as plt
plt.title('Popularity')
sea.set(rc={'figure.figsize': (20, 20)})
sea.lineplot(data = dataFrame1['Despacito'])
sea.lineplot(data = dataFrame1['Shape of You'])
plt.xlabel('sample x')
```

```python
flight_data = pd.read_csv('https://raw.githubusercontent.com/Laxminarayen/Inceptez_batch_16/main/flight_delays.csv')
plt.figure(figsize= (16,6))
plt.title("Average Arrival delay for Inceptez Airlines WN, by month")

#defenition for the bar 
sea.barplot(x = flight_data.index, y = flight_data['WN'])

#Adding labels 
plt.ylabel("Arrival delay in mins")
plt.xlabel("The months as integers")
```

```python
plt.figure(figsize= (16, 6))
plt.title("Average Arrival delay for Inceptez Airlines WN, by month")

#defenition for the bar 
sea.heatmap(data = flight_data, annot=True)

#Adding labels 
plt.ylabel("Arrival delay in mins")
plt.xlabel("The months as integers")
```

```python
plt.figure(figsize= (16, 6))
plt.title("Average Arrival delay for Inceptez Airlines WN, by month")

#defenition for the bar 
sea.heatmap(data = flight_data.corr(), annot=True, vmin= -1, vmax = 1)

#Adding labels 
plt.ylabel("Arrival delay in mins")
plt.xlabel("The months as integers")
```

```python
insurance = pd.read_csv('https://raw.githubusercontent.com/Laxminarayen/Inceptez_batch_16/main/insurance.csv')
```

```python
insurance.head()
```

```python
plt.figure(figsize= (6, 6))
sea.scatterplot(x = insurance['age'], y = insurance['charges'], hue = insurance['smoker'])
```

```python
plt.figure(figsize= (6, 6))
sea.scatterplot(x = insurance['bmi'], y = insurance['charges'], hue = insurance['smoker'])
```

```python
plt.figure(figsize= (6, 6))
sea.regplot(x = insurance['bmi'], y = insurance['charges'])
```

```python
plt.figure(figsize= (6, 6))
sea.lmplot(x = 'bmi', y = 'charges', hue = 'smoker', data = insurance)
```

```python
plt.figure(figsize= (6, 6))
sea.swarmplot(x = insurance['smoker'], y = insurance['charges'])
```

```python
plt.figure(figsize= (6, 6))
sea.stripplot(x = insurance['smoker'], y = insurance['charges'])
```
